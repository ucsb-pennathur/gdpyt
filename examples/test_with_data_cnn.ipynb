```python
from gdpyt.utils import generate_sig_settings, generate_grid_calibration, generate_synthetic_images, plot_tensor_dset
from gdpyt.utils.nn import GdpytTensorDataset, BCAdjust
from gdpyt import GdpytCalibrationSet, GdpytImageCollection
import matplotlib.pyplot as plt
import numpy as np
from os.path import join
%matplotlib qt
```

Load images to be analyzed

```python
filetype = '.tif'
processing = {
    'cv2.convertScaleAbs': {'args': [], 'kwargs': dict(alpha=0.5, beta=0)},
    #'cv2.GaussianBlur': {'args': [(5, 5), 5]},
    'cv2.medianBlur': {'args': [5]}
}
# folder = r'C:\Users\silus\UCSB\master_thesis\python_stuff\gdpyt\tests\test_data\chip2test1\50X\calib'
folder = r'C:\Users\silus\UCSB\master_thesis\python_stuff\gdpyt\tests\test_data\chip2test1\50X\test\chip1_E20_P2_f0.25_Obj50X_FPNR5_run_22'
collection = GdpytImageCollection(folder, filetype, processing_specs=processing,
                                 min_particle_size=20, max_particle_size=5000, shape_tol=0.2)
collection.uniformize_particle_ids(threshold=50)
```

```python
fig = collection.plot(raw=True, draw_particles=False)
fig.show()
```

```python
# Make example tensor dataset just to look at it
chip2_dset = GdpytTensorDataset(normalize=False)
chip2_dset.from_image_collection(collection, ref_shape=(44, 44), max_size=50)
fig = plot_tensor_dset(chip2_dset, 20)
fig.show()
```

Generate calibration images set

```python
# SKIP THIS CELL IF YOU WANT TO USE EXPERIMENTAL CALIB STACKS ##########################################################
# Background intensity and noise of the collection to be analyzed
mean = []
sigma = []
for i in range(len(collection)):
    stats = collection[i].stats[['mean_bckgr_r', 'sigma_bckgr_r']].values.squeeze()
    mean.append(stats[0])
    sigma.append(stats[1])
    
fig, ax = plt.subplots(nrows=2)
ax[0].hist(mean)
ax[1].hist(sigma)
```

```python
# SKIP THIS CELL IF YOU WANT TO USE EXPERIMENTAL CALIB STACKS ##########################################################
setup_params = dict(
    magnification = 50,
    numerical_aperture = 0.5,
    focal_length = 2500,
    ri_medium = 1,
    ri_lens = 1.5,
    pixel_size = 16,
    pixel_dim_x = 512,
    pixel_dim_y = 512,
    background_mean = 200,
    background_noise = 0,
    points_per_pixel = 18,
    n_rays = 100,
    gain = 1,
    cyl_focal_length = 0
)

synthetic_folder = r'C:\Users\silus\UCSB\master_thesis\python_stuff\gdpyt\tests\test_synthetic\Chip2Test1_50x_Synthetic_f2500'
settings_dict, settings_path = generate_sig_settings(setup_params,
                                                     folder=synthetic_folder)
grid = (4, 4)
generate_grid_calibration(settings_path, grid, np.linspace(-30, 30, 61), particle_diameter=5, create_multiple=5)
#generate_grid_calibration(settings_path, grid, [0, 5], particle_diameter=5, create_multiple=None)
generate_synthetic_images(settings_path, join(synthetic_folder, 'calibration_input'), join(synthetic_folder, 'calibration_images'))
```

```python
# SKIP THIS CELL IF YOU WANT TO USE EXPERIMENTAL CALIB STACKS ##########################################################
folder = join(synthetic_folder, 'calibration_images')
processing = {
    'cv2.convertScaleAbs': {'args': [], 'kwargs': dict(alpha=0.7, beta=0)},
    #'cv2.GaussianBlur': {'args': [(21, 21), 5]}
    'cv2.medianBlur': {'args': [5]}
}
synth_collection = GdpytImageCollection(folder, filetype, processing_specs=processing,
                                 min_particle_size=30, max_particle_size=5000, shape_tol=0.2)
synth_collection.uniformize_particle_ids(threshold=30)
name_to_z = {}
for image in synth_collection.images.values():
    name_to_z.update({image.filename: float(image.filename.split('_')[-1].split(filetype)[0])})

calib_set = synth_collection.create_calibration(name_to_z)
```

```python
# SKIP THIS CELL IF YOU WANT TO USE EXPERIMENTAL CALIB STACKS ##########################################################
fig = synth_collection.plot(raw=False, draw_particles=True)
fig.show()
```

```python
folder_2 = r'C:\Users\silus\UCSB\master_thesis\python_stuff\gdpyt\tests\test_data\0319_loc1\50X\0319_calib'
exclude = ['calib_{}.tif'.format(int(x)) for x in range(41, 56)]
calib_2 = GdpytImageCollection(folder_2, filetype, processing_specs=processing,
                                        min_particle_size=30, max_particle_size=5000, shape_tol=0.9, exclude=exclude)
calib_2.uniformize_particle_ids(threshold=30)
name_to_z_2 = {}
for image in calib_2.images.values():
    name_to_z_2.update({image.filename: float(image.filename.split('_')[-1].split('.')[0])})
```

Make tensor dataset to see what the samples will look like

```python
#bc_adjust = BCAdjust(0.5, 75)
synth_dset = GdpytTensorDataset(normalize=False, transforms_=None)
synth_dset.from_calib_set(calib_set, max_size=100)#_collection(synth_collection, ref_shape=(44, 44), max_size=100)
fig = plot_tensor_dset(synth_dset, 20)
fig.show()
```

Training CNN

```python
# Create and train CNN #######################################################################################
calib_set.create_cnn(n_conv_layers=3, n_linear_layers=2, kernel_size=11, n_filters_init=2, batch_norm=(0, 1, 2))
epochs = 10
lr = 5e-4
transforms = None #[BCAdjust(0.5, 75)] #[RotateN90(), RandomBCGAdjust()]
calib_set.train_cnn(epochs, 'nn.MSELoss()', normalize_inputs=True, transforms=transforms, 
                    max_sample_size=100, skip_na=True, reg_type='l2', batch_size=64)

fig, ax = plt.subplots(figsize=(7, 5))
ax.errorbar(calib_set.train_summary['Epoch'], calib_set.train_summary['Avg_loss'],
            yerr=calib_set.train_summary['Sigma_loss'], capsize=2)
ax.set_ylabel('Error (L1)')
ax.set_xlabel('Epoch')
ax.grid(True)
ax.set_yscale('log')
```
